### Train PD Model : Logistic Regression Model with P-Values

```python
# P values for sklearn logistic regression.#
# Class to display p-values for logistic regression in sklearn.
class LogisticRegression_with_p_values:
    
    def __init__(self,*args,**kwargs):#,**kwargs):
        self.model = linear_model.LogisticRegression(*args,**kwargs)#,**args)

    def fit(self,X,y):
        self.model.fit(X,y)
        
        #### Get p-values for the fitted model ####
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom,(X.shape[1],1)).T
        F_ij = np.dot((X / denom).T,X) ## Fisher Information Matrix
        Cramer_Rao = np.linalg.inv(F_ij) ## Inverse Information Matrix
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.model.coef_[0] / sigma_estimates # z-score for eaach model coefficient
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores] ### two tailed test for p-values
        
        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_
        self.p_values = p_values
```

```python
plreg = LogisticRegression_with_p_values(solver='lbfgs', max_iter=200)
plreg.fit(X_train, y_train)
```

- #### Logistic Regression Model Evaluation

```python
y_hat = logreg.predict(X_test)
y_hat
```

This is how the method `.predict()` works behind the scenes:
1. First, the predict method multiplies the values of the variables of each observation by the estimated model coefficients. This yields the log of odds and non default
2. Then an exponent is raised to the power of this log odds and thus the odds themselves are obtained. From there, the estimated probability of non default is obtained.
3. Finally, the results are categorized into non default and default by applying a cutoff to this probability. The default cutoff is 50% percent. 

We can also get the predicted probabilities instead of the classes :

```python
y_hat_prob = logreg.predict_proba(X_test)
y_hat_prob
```

```python
## Create a nice format for the prediction
y_hat_prob =  y_hat_prob[:][:,1]
y_test = pd.read_csv('target_test.csv', index_col = 0)
y_test_temp = y_test
y_test_temp.reset_index(drop = True, inplace = True)

pred = pd.concat([y_test_temp, pd.DataFrame(y_hat_prob)], axis = 1)
pred.columns = ['y_test', 'y_hat_prob']
pred.index = X_test.index

# Set up threshold
tr = 0.5
pred['y_hat'] = np.where(pred['y_hat_prob'] > tr, 1, 0)
pred.head()
```

For convinience of creating scorecard, we need the summary table also contain the reference categories
and a new column contain original feature name. The code below accompplish all of this.


```python
# Delete unecessary column 
summary_table.drop(columns=['p_value<0.05'], inplace=True)

# List of reference categories
ref_cats = [
     'branch_code:E',
     'delinquency_score:1-5', 
     'number_of_cards:1-3', 
     'payment_ratio:>0_and_<=10',
     'overlimit_percentage:>22.706',
     'payment_ratio_3month:<0',
     'payment_ratio_6month:<0',
     'remaining_bill_per_limit:>1.217',
     'total_usage_per_limit:>0.2',
     'total_3mo_usage_per_limit:>0.0186',
     'utilization_3month:>0.975', 
     'utilization_6month:>0.938'
]
# Create dataframe to be concated with summary_table
df_ref_cats = pd.DataFrame(ref_cats, columns = ['Feature name'])
df_ref_cats['Coefficients'] = 0
df_ref_cats['p_values'] = np.nan

# Concat the two dataframe
df_scorecard = pd.concat([summary_table, df_ref_cats])
df_scorecard = df_scorecard.reset_index()

# Create a new column, called 'Original feature name', 
# which contains the value of the 'Feature name' column, up to the column symbol.
df_scorecard['Original feature name'] = df_scorecard['Feature name'].str.split(':').str[0]

df_scorecard
```

To create a scorecard, we need to turn the regression coefficients from our PD model into simple scores.
So we must decide what scorecard we want to create. Let's say we want the minimum score to be 300 and the maximum score to be 850.

```python
min_score = 300
max_score = 850
```

So we also need the minimum and the maximum of the credit worthiness assessments that can be obtained from the PD model. From the interpretation of the coefficient of logistic regression model, higher coefficients correspond to being a better credit worthiness. _So, the minimum credit worthiness assessment we can get from the model would
be in the case where a user falls into all the categories of original independent variables with lowest model coefficients. Similarly for the maximum counterparts_.

Let's find this values using groupby method.

```python
print(df_scorecard.groupby('Original feature name')['Coefficients'].min()) 
print()
print(df_scorecard.groupby('Original feature name')['Coefficients'].max())
```

#### Sum of Coefficients to Credit Score 

What we really want is a simple linear rule (formula) to compute a scorecard given a user data (represented by a series of 1 or 0 number indicating values of dummy variables)

$$
\text{score} = \text{intercept_score} + \sum_{\text{variable}} \{1 \text{ or } 0\} \, * \, \text{variable_score} 
$$

Now with maximum and minimum scores of Scorecard already determined, each variable score computed as follows

$$
\text{variable_score} = \text{variable_coef} \frac{(\text{max_score} - \text{min_score})}{(\text{max_sum_coef} - \text{min_sum_coef})} 
$$
and the intercept score determined as
$$
\text{intercept_score} = (\text{intercept_coef} - \text{min_score}) \frac{(\text{max_score} - \text{min_score})}{(\text{max_sum_coef} - \text{min_sum_coef})} + \text{min_score}
$$

```python
min_sum_coef = df_scorecard.groupby('Original feature name')['Coefficients'].min().sum()
max_sum_coef = df_scorecard.groupby('Original feature name')['Coefficients'].max().sum()

# Calculate variable scores and intercept score
df_scorecard['Score - Calculation'] = df_scorecard['Coefficients'] * (max_score - min_score) / (max_sum_coef - min_sum_coef)
df_scorecard['Score - Calculation'][0] = ((df_scorecard['Coefficients'][0] - min_sum_coef) / (max_sum_coef - min_sum_coef)) * (max_score - min_score) + min_score

# Rounding to make it easier to use
df_scorecard['Score - Preliminary'] = df_scorecard['Score - Calculation'].round()
df_scorecard['Difference'] = df_scorecard['Score - Preliminary'] - df_scorecard['Score - Calculation']

# min_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Preliminary'].min().sum()
# max_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Preliminary'].max().sum()

df_scorecard['Score - Final'] = df_scorecard['Score - Preliminary']
df_scorecard['Score - Final'][5] = 25 # Compensate for rounding

# min_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Final'].min().sum()
# max_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Final'].max().sum()

df_scorecard
```

## Calculating Credit Scores
To calculate the credit score for all user all at once (without for loop), we need to modify the test set `X_test_p_val`. We basically want to compute 
$$
\text{score} = \text{intercept_score} + \sum_{\text{variable}} \{1 \text{ or } 0\} \, * \, \text{variable_score} 
$$
for each user (each row in test set). When doing it all at once for all user we basically doing matrix multiplication (the sum term) of `X_test_p_val` with `df_scorecard['Score - Final']` and addition (with a broadcasted constant, the intercept value). Let's check the dimensions first.

```python
print(X_test_p_val.shape)
print()
print(df_scorecard['Score - Final'].shape)
```

```python
# So we have to reshape `df_scorecard['Score - Final']` and take out the reference categories
score_var = df_scorecard['Score - Final'][1:39]
score_var = score_var.values.reshape(38,1)
score_var.shape
```

```python
# Calculate the credit score
y_test_scores = X_test_p_val.dot(score_var) + df_scorecard['Score - Final'][0]
y_test_scores.rename(columns={0: "credit_score"}, inplace=True)
pd.concat([y_test_scores,pd.read_csv('target_test.csv', index_col = 0)], axis=1)
```

## Setting Cut-off
Recall that from [model evaluation section](#Logistic-Regression-Model-with-p-value-Evaluation) we already computed : tpr, fpr and thresholds as outputs from `roc_curve()` function, stored in variables `tpr2, fpr2, thresholds2`. Let's store them in a dataframe :

```python
df_cutoffs = pd.concat([pd.DataFrame(thresholds2), pd.DataFrame(fpr2), pd.DataFrame(tpr2)], axis = 1)
df_cutoffs.columns = ['thresholds', 'fpr', 'tpr']
df_cutoffs
```

```python
# The threhold in the first row doesn't make sense since it is higher than 1, as maximum probability. 
# Let's make it really close to 1 instead 
df_cutoffs['thresholds'][0] = 1 - 1 / np.power(10, 16)
```

When creating scorecard, we use the following formulas to get the score from sum of coefficients

$$
\text{score} = \text{intercept_score} + \sum_{\text{variable}} \{1 \text{ or } 0\} \, \cdot \, \text{variable_score}.
$$
Now we can go backwards, from score to sum of coeficients, by formula

$$
\text{sum_of_coef} = \text{min_sum_coef} + (\text{score} - \text{min_score}) \cdot 
\frac{(\text{max_sum_coef} - \text{min_sum_coef})}{(\text{max_score} - \text{min_score})}
$$

Furthermore, because probability of non-default obtained from sum of coefficients by logistic function,

$$
\ln \Big( \frac{\text{prob_non-default}}{\text{prob_default}} \Big) = \beta_0 + \beta_1 X_1 + \cdots + \beta_m X_m = \text{sum_of_coef}, 
$$
we then have the relationship to obtained score from probability of non-default as follows

$$
\text{score_from_prob} = \text{min_score} + \Bigg( \ln \frac{\text{prob_non-default}}{\text{prob_default}}- \text{min_sum_coef} \Bigg)  \cdot   \frac{(\text{max_score} - \text{min_score})}{(\text{max_sum_coef} - \text{min_sum_coef})}.
$$

_Since probability of non-default determine the default rate, we can compute or obtain optimal credit score cut-off from a prescribed default rate by above formula_. Below is the implementation: 

```python
# Create score column correspond to threshold probability using the last formula
df_cutoffs['Score'] = ((np.log(df_cutoffs['thresholds'] / (1 - df_cutoffs['thresholds'])) - min_sum_coef) 
                       * ((max_score - min_score) / (max_sum_coef - min_sum_coef)) + min_score).round()
df_cutoffs.head(10)
```

```python
# The first eight rows doesn't look nice since the scores already higher than maximum score 850.
# Let's remove them and modify the first row to be the maximum score
df_cutoffs.drop(df_cutoffs.index[1:8], inplace=True)
df_cutoffs.reset_index(inplace=True, drop=True)
df_cutoffs['Score'][0] = max_score
df_cutoffs.head(10)
```

```python
def n_approved(p):
    """
    We define a function called 'n_approved' which assigns a value of 1 if a predicted probability
    is greater than the parameter p, which is a threshold, and a value of 0, if it is not.
    Then it sums the column.
    Thus, if given any percentage values, the function will return
    the number of rows wih estimated probabilites greater than the threshold.
    """
    return np.where(pred2['y_hat2_prob'] >= p, 1, 0).sum()

# Now let's compute the default and non-default rate for each possible threshold (probs cut-off)
df_cutoffs['N non-default'] = df_cutoffs['thresholds'].apply(n_approved)
df_cutoffs['N default'] = pred2['y_hat2_prob'].shape[0] - df_cutoffs['N non-default']
df_cutoffs['Non-default Rate'] = df_cutoffs['N non-default'] / pred2['y_hat2_prob'].shape[0]
df_cutoffs['Default Rate'] = 1 - df_cutoffs['Non-default Rate']
```
