### Missing Values and Duplicate Data

```python
Desc = []

for i in df.columns:
    Desc.append([
        i,
        df[i].dtypes,
        df[i].isna().sum(),
        round((((df[i].isna().sum() )/ len(df)) * 100),2),
        df[i].nunique(),
        df[i].drop_duplicates().sample(2).values
    ])

dM = pd.DataFrame(data = Desc,
            columns = [
                'Column',
                'DataType',
                'Null',
                'Null (%)',
                'nUnique',
                'Unique Samples'
            ])
dM
```

Columns with missing values are :

```python
dM[dM['Null']!=0] # Features with missing values
```

#### Duplicate Data

```python
# Data duplication based on all columns
df[df.duplicated()]
```

```python
# Data duplication based on all columns except user id column (X column)
columns_list_without_X = df.drop(columns='X').columns.to_list()
dy = df[df.duplicated(subset=columns_list_without_X, keep=False)]
dy
```

So, without consider the X column (user_id), there are duplicate rows. For example these two rows below :

```python
dy[dy['outstanding'] == 5972272]
```

However, these rows are not entirely duplicates, since the branch_code values on both rows are missing. Indeed this happens on all the 'duplicate' rows :  

```python
dy['branch_code'].unique()
```

```python
dy.info()
```

So, all the duplicate rows have missing values on branch_code column, but all other columns have zero missing value. This raise a question : "Is all rows in the original dataframe with missing branch_code value is actually the duplicates rows above?" The answer is no. Since the total number of rows are different. We have $188$ duplicate rows (rows in `dy` dataframe), which is less than the total number of rows with missing branch_code value  as shown below: 

```python
df['branch_code'].isnull().sum()
```

These 'duplicates' rows are quite suspicious. I won't drop them for now because they're not really duplicate since `branch_code` values are missing. Besides, the user_id for them are different. I'll analyse them later

### Outliers and Statistical Summary

```python
# Outliers by Tabular
Desc_outliers = []

for i in df.drop(columns = ['X', 'branch_code', 'default_flag']):
    Q1 = df.describe()[i]['25%']
    Q3 = df.describe()[i]['75%']
    IQR = Q3 - Q1
    upper_fence = Q3 + (1.5*IQR)
    lower_fence = Q1 - (1.5*IQR)
    hitung = 0
    total = 0
    
    for j in df[i]:
        if (j < lower_fence) | (j > upper_fence) :
            hitung+=1
        if np.isnan(j) == False:
            total+=1
    Desc_outliers.append([
        i,
        hitung,
        round((hitung/total)*100,3)
    ])
            
# upper_fence, lower_fence
pd.DataFrame(Desc_outliers, columns=[
    "Feature",
    "number_of_Outliers",
    "%outliers"
])
```
