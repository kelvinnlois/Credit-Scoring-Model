## Data Cleaning, Splitting Data and Imputation

````python
## Change the data type of outstanding column to float
df[['outstanding']] = df[['outstanding']].astype(float)

## Change the datatype of deliquency_score column to object
df['delinquency_score'] = df['delinquency_score'].fillna(-1)
df['delinquency_score'] = df['delinquency_score'].astype(int)
df['delinquency_score'] = df['delinquency_score'].astype(str)
df['delinquency_score'] = df['delinquency_score'].replace('-1', np.nan)

## Change target column (default_flag) for convenience
df['non-default_flag'] = df['default_flag'].map({0:1, 1:0})
df.drop('default_flag', axis = 1, inplace=True)
````

#### Splitting Data

Takes a set of inputs and a set of targets as arguments. Splits the inputs and the targets into four dataframes:
Inputs - Train, Inputs - Test, Targets - Train, Targets - Test.

We split two dataframes with inputs and targets, each into a train and test dataframe, and store them in variables.

````python
# Splitting Data into train and test set
X_train, X_test, y_train, y_test = train_test_split(df.drop('non-default_flag', axis = 1), 
                                                    df['non-default_flag'],
                                                    test_size = 0.2, 
                                                    random_state = 42)
print("X_train shape \t: ", X_train.shape, "\n"
      "X_test shape \t: ", X_test.shape, "\n"
      "y_train shape \t: ", y_train.shape, "\n"
      "y_test shape \t: ", y_test.shape
     )
````

### Preprocessing Unordered Discrete Variables

- #### WoE functions for Unordered Discrete Variables
In our case, this is WoE function for branch_code variable, the only unordered discrete variable in our data. 

````python
# WoE function for unordered discrete variables to compute WoE and IV
def woe_cat(df_input, cat_variabe_name, df_target):
    """
    The function takes 3 arguments: a dataframe input, a string (the name of the categorical variable), 
    and a dataframe of target. The function returns a dataframe contain woe and information value columns.     
    """
    df_input = pd.concat([df_input[cat_variabe_name], df_target], axis = 1)
    df_input = pd.concat([df_input.groupby(df_input.columns.values[0], as_index = False)[df_input.columns.values[1]].count(),
                          df_input.groupby(df_input.columns.values[0], as_index = False)[df_input.columns.values[1]].mean()], axis = 1)
    df_input = df_input.iloc[:, [0, 1, 3]]
    df_input.columns = [df_input.columns.values[0], 'n_obs', 'prop_good']
    df_input['prop_n_obs'] = df_input['n_obs'] / df_input['n_obs'].sum()
    df_input['n_good'] = df_input['prop_good'] * df_input['n_obs']
    df_input['n_bad'] = (1 - df_input['prop_good']) * df_input['n_obs']
    df_input['prop_n_good'] = df_input['n_good'] / df_input['n_good'].sum()
    df_input['prop_n_bad'] = df_input['n_bad'] / df_input['n_bad'].sum()
    df_input['WoE'] = np.log(df_input['prop_n_good'] / df_input['prop_n_bad'])
    df_input = df_input.sort_values(['WoE'])
    df_input = df_input.reset_index(drop = True)
    df_input['diff_prop_good'] = df_input['prop_good'].diff().abs()
    df_input['diff_WoE'] = df_input['WoE'].diff().abs()
    df_input['IV'] = (df_input['prop_n_good'] - df_input['prop_n_bad']) * df_input['WoE']
    df_input['IV'] = df_input['IV'].sum()
    return df_input

# Function for Visualize the result/output of function woe_cat : Plot WoE vs cat variable
def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
    """
    The function that takes 2 arguments: a dataframe (output of woe function) and a number.
    The function displays a graph WoE vs cat variable.
    """
    x = np.array(df_WoE.iloc[:, 0].apply(str))
    # Turns the values of the column with index 0 to strings, makes an array from these strings, and passes it to variable x.
    y = df_WoE['WoE']
    # Selects a column with label 'WoE' and passes it to variable y.
    plt.figure(figsize=(18, 6))
    # Sets the graph size to width 18 x height 6.
    plt.plot(x, y, marker = 'o', linestyle = '--', color = 'k')
    # Plots the datapoints with coordiantes variable x on the x-axis and variable y on the y-axis.
    # Sets the marker for each datapoint to a circle, the style line between the points to dashed, and the color to black.
    plt.xlabel(df_WoE.columns[0])
    # Names the x-axis with the name of the column with index 0.
    plt.ylabel('Weight of Evidence')
    # Names the y-axis 'Weight of Evidence'.
    plt.title(str('Weight of Evidence by ' + df_WoE.columns[0]))
    # Names the grapth 'Weight of Evidence by ' the name of the column with index 0.
    plt.xticks(rotation = rotation_of_x_axis_labels)
    # Rotates the labels of the x-axis a predefined number of degrees.
````

### Preprocessing Continuous and Ordered Discrete Variables
We will create WoE function to create coarse classes of our ordered discrete variable (delinquency_score)
and our continuous variables. Since order matter, we need to make a slight modification of our WoE function for unordered discrete variables before. We will preserve the natural order of the fine classes, instead of
reordered the classes by their WoE values.

- #### WoE functions for Ordered Discrete Variables and Continuous Variables


````python
# WoE function for ordered discrete variables and continuous variables
def woe_ordered_cont(df_input, cat_variabe_name, df_target):
    """
    The function takes 3 arguments: a dataframe input, a string (the name of the categorical variable), 
    and a dataframe of target. The function returns a dataframe contain woe and information value columns.     
    """
    df_input = pd.concat([df_input[cat_variabe_name], df_target], axis = 1)
    df_input = pd.concat([df_input.groupby(df_input.columns.values[0], as_index = False)[df_input.columns.values[1]].count(),
                          df_input.groupby(df_input.columns.values[0], as_index = False)[df_input.columns.values[1]].mean()], axis = 1)
    df_input = df_input.iloc[:, [0, 1, 3]]
    df_input.columns = [df_input.columns.values[0], 'n_obs', 'prop_good']
    df_input['prop_n_obs'] = df_input['n_obs'] / df_input['n_obs'].sum()
    df_input['n_good'] = df_input['prop_good'] * df_input['n_obs']
    df_input['n_bad'] = (1 - df_input['prop_good']) * df_input['n_obs']
    df_input['prop_n_good'] = df_input['n_good'] / df_input['n_good'].sum()
    df_input['prop_n_bad'] = df_input['n_bad'] / df_input['n_bad'].sum()
    df_input['WoE'] = np.log(df_input['prop_n_good'] / df_input['prop_n_bad'])
#     df_input = df_input.sort_values(['WoE'])
#     df_input = df_input.reset_index(drop = True)
    df_input['diff_prop_good'] = df_input['prop_good'].diff().abs()
    df_input['diff_WoE'] = df_input['WoE'].diff().abs()
    df_input['IV'] = (df_input['prop_n_good'] - df_input['prop_n_bad']) * df_input['WoE']
    df_input['IV'] = df_input['IV'].sum()
    return df_input
````

- #### Preprocessing credit_limit variable
First, we do fine classing (or categorize) for this continuous feature by applying `pd.cut()` as follows. Say that we want 50 categories :   

````python
X_train_prep['credit_limit_factor'] = pd.cut(X_train_prep['credit_limit'], 50)
````

````python
df_temp = woe_ordered_cont(X_train_prep, 'credit_limit_factor', y_train)
df_temp
````

Here we see that 
- there's a very large number of observation for the first interval/category, and 
- a lot of empty observations for large intervals. 

If we try to increase the number of categories to, say 100, it is still not enough to get a good proportions to other intervals (and the proportion of the first intervals is still 60%). Besides, the number of large intervals with empty observations will also significantly increased. So we need another strategy to deal with this. Let's explore the following strategy :
Based on the 50 intervals given above,
1. Create an interval (a category) that contain large credit limits intervals, with a chosen threshold as lower bound : $(\text{threshold}, 1000000000.0]$. The threshold is better chosen with the help of domain knowledge, so it is a lower bound for high credit limits. By looking at the dataframe above, the good candidate maybe $42880000.0$, that is the lower bound of the third interval $(42880000.0, 62820000.0]$.
2. Cut the credit_limit variable into  50 categories but only for data with credit limit less than the threshold.

````python
X_temp = X_train_prep.loc[X_train_prep['credit_limit'] <= 42880000.0, : ]
X_temp["credit_limit_factor"] = pd.cut(X_temp['credit_limit'], 30)
df_temp = woe_ordered_cont(X_temp, 'credit_limit_factor', y_train[X_temp.index])
df_temp
````

````python
plot_by_woe(df_temp, 90)
````

This plot shows that the WoE vary wildly with credit_limit_factor. These variation following the trend that almost horizontal. So there's no clear pattern of dependency between the variable with weight of evidence. __This means that it might be a decision if we didn't use this variable (credit_limit) in our PD model__.
